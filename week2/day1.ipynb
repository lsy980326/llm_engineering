````python
# 임포트

import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')
grok_api_key = os.getenv('GROK_API_KEY')
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if openai_api_key:
    print(f"OpenAI API 키 있음, 접두사: {openai_api_key[:8]}")
else:
    print("OpenAI API 키 없음")
    
if anthropic_api_key:
    print(f"Anthropic API 키 있음, 접두사: {anthropic_api_key[:7]}")
else:
    print("Anthropic API 키 없음 (선택 사항)")

if google_api_key:
    print(f"Google API 키 있음, 접두사: {google_api_key[:2]}")
else:
    print("Google API 키 없음 (선택 사항)")

if deepseek_api_key:
    print(f"DeepSeek API 키 있음, 접두사: {deepseek_api_key[:3]}")
else:
    print("DeepSeek API 키 없음 (선택 사항)")

if groq_api_key:
    print(f"Groq API 키 있음, 접두사: {groq_api_key[:4]}")
else:
    print("Groq API 키 없음 (선택 사항)")

if grok_api_key:
    print(f"Grok API 키 있음, 접두사: {grok_api_key[:4]}")
else:
    print("Grok API 키 없음 (선택 사항)")

if openrouter_api_key:
    print(f"OpenRouter API 키 있음, 접두사: {openrouter_api_key[:3]}")
else:
    print("OpenRouter API 키 없음 (선택 사항)")
````

````python
# OpenAI 클라이언트 라이브러리 연결
# HTTP 엔드포인트 호출을 감싼 얇은 래퍼

openai = OpenAI()

# Gemini, DeepSeek, Groq는 OpenAI Python 클라이언트로 호출 가능
# Google·DeepSeek이 OpenAI 호환 엔드포인트를 제공하고, base_url을 바꿀 수 있기 때문

anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
deepseek_url = "https://api.deepseek.com"
groq_url = "https://api.groq.com/openai/v1"
grok_url = "https://api.x.ai/v1"
openrouter_url = "https://openrouter.ai/api/v1"
ollama_url = "http://localhost:11434/v1"

anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
deepseek = OpenAI(api_key=deepseek_api_key, base_url=deepseek_url)
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
grok = OpenAI(api_key=grok_api_key, base_url=grok_url)
openrouter = OpenAI(base_url=openrouter_url, api_key=openrouter_api_key)
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
````

## 이제 재미로 — 챗봇끼리 대립 대화하기

이미 프롬프트가 리스트로 구성되는 것은 익숙하실 겁니다:

```
[
    {"role": "system", "content": "시스템 메시지"},
    {"role": "user", "content": "사용자 프롬프트"}
]
```

이 구조로 **대화 이력**을 이렇게 쌓을 수 있습니다:

```
[
    {"role": "system", "content": "시스템 메시지"},
    {"role": "user", "content": "첫 사용자 메시지"},
    {"role": "assistant", "content": "어시스턴트 응답"},
    {"role": "user", "content": "새 사용자 메시지"},
]
```

이 방식으로 이력을 유지한 긴 대화를 만들 수 있습니다.

````python
# GPT-4.1-mini와 Claude-3.5-haiku가 대화하도록 만들기
# 비용을 줄이기 위해 저렴한 모델 버전 사용

gpt_model = "gpt-4.1-mini"
gemini_model = "gemini-2.5-flash-lite"

gpt_system = "당신은 말싸움을 좋아하는 챗봇이다. 대화 내용에 반박하고 모든 걸 도전적으로 비판한다. 비꼬는 말투로."
gemini_system = "당신은 매우 정중한 챗봇이다. 상대 말에 맞장구치거나 공통점을 찾는다. 상대가 공격적이면 달래 주며 대화를 이어간다."

gpt_messages = ["안녕"]
gemini_messages = ["안녕, 반가워"]
````

````python
def call_gpt():
    messages = [{"role": "system", "content": gpt_system}]
    for gpt, gmi in zip(gpt_messages, gemini_messages):
        messages.append({"role": "assistant", "content": gpt})
        messages.append({"role": "user", "content": gmi})
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_gpt()
````

````python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]
    for gpt, gmi in zip(gpt_messages, gemini_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": gmi})
    messages.append({"role": "user", "content": gpt_messages[-1]})
    response = gemini.chat.completions.create(model=gemini_model, messages=messages)
    return response.choices[0].message.content
````
