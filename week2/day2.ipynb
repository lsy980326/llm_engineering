```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
import gradio as gr  # 등장!
```

```python
# .env 파일에서 환경 변수 로드
# 디버깅용으로 키 접두사 출력
# 원하는 제공자만 써도 되고, 전부 Ollama만 써도 됨

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')

if openai_api_key:
    print(f"OpenAI API 키 있음, 접두사: {openai_api_key[:8]}")
else:
    print("OpenAI API 키 없음")
    
if anthropic_api_key:
    print(f"Anthropic API 키 있음, 접두사: {anthropic_api_key[:7]}")
else:
    print("Anthropic API 키 없음")

if google_api_key:
    print(f"Google API 키 있음, 접두사: {google_api_key[:8]}")
else:
    print("Google API 키 없음")
```

```python
# OpenAI, Anthropic, Google 연결 — 사용 안 하면 해당 줄 주석 처리

openai = OpenAI()

# anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"

# anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
```

```python
# GPT-4.1-mini 호출을 간단한 함수로 감싸기

system_message = "당신은 도움이 되는 어시스턴트입니다"

def message_gpt(prompt):
    messages = [{"role": "system", "content": system_message}, {"role": "user", "content": prompt}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
# "학습 컷오프" 즉 학습 데이터의 최신 시점을 확인할 수 있음

message_gpt("오늘 날짜가 뭐야?")
```

```python
# 마크다운 출력 사용
# system_message를 아래 코드에서 안 쓰는데 왜 바꾸나요? message_gpt 함수 안에서 전역 변수로 쓰이기 때문입니다 (한번 확인해 보세요)
# 좋은 소프트웨어 관행은 아니지만, Jupyter 랩에서 R&D할 때는 흔한 패턴

system_message = "코드 블록 없이 마크다운으로만 답하는 도움 되는 어시스턴트"

message_input = gr.Textbox(label="메시지:", info="GPT-4.1-mini에 보낼 내용", lines=7)
message_output = gr.Markdown(label="응답:")

view = gr.Interface(
    fn=message_gpt,
    title="GPT", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "일반인에게 트랜스포머 아키텍처 설명해 줘",
        "AI 엔지니어 지망생에게 트랜스포머 아키텍처 설명해 줘",
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
# 스트리밍으로 결과를 받는 호출 만들기
# 제너레이터(yield) 복습이 필요하면 guides 폴더의 Intermediate Python 가이드를 보세요

def stream_gpt(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = openai.chat.completions.create(
        model='gpt-4.1-mini',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="메시지:", info="GPT-4.1-mini에 보낼 내용", lines=7)
message_output = gr.Markdown(label="응답:")

view = gr.Interface(
    fn=stream_gpt,
    title="GPT", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "일반인에게 트랜스포머 아키텍처 설명해 줘",
        "AI 엔지니어 지망생에게 트랜스포머 아키텍처 설명해 줘",
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_gemini(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = gemini.chat.completions.create(
        model='gemini-2.5-flash',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="메시지:", info="gemini에 보낼 내용", lines=7)
message_output = gr.Markdown(label="응답:")

view = gr.Interface(
    fn=stream_gemini,
    title="gemini", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "일반인에게 트랜스포머 아키텍처 설명해 줘",
        "AI 엔지니어 지망생에게 트랜스포머 아키텍처 설명해 줘",
        ], 
    flagging_mode="never"
    )
view.launch()
```

## 이제 조금 더 다듬기

제너레이터와 "yield"가 헷갈리면 Intermediate Python 가이드를 다시 보세요.

```python
def stream_model(prompt, model):
    if model=="GPT":
        result = stream_gpt(prompt)
    elif model=="gemini":
        result = stream_gemini(prompt)
    else:
        raise ValueError("알 수 없는 모델")
    yield from result
```

```python
message_input = gr.Textbox(label="메시지:", info="LLM에 보낼 내용", lines=7)
model_selector = gr.Dropdown(["GPT", "gemini"], label="모델 선택", value="GPT")
message_output = gr.Markdown(label="응답:")

view = gr.Interface(
    fn=stream_model,
    title="LLMs", 
    inputs=[message_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["일반인에게 트랜스포머 아키텍처 설명해 줘", "GPT"],
            ["AI 엔지니어 지망생에게 트랜스포머 아키텍처 설명해 줘", "gemini"]
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
# 실험할 때 자주 쓰는 방식 — 위에서 쓰던 전역 변수를 바꿈

system_message = """
회사 웹사이트 랜딩 페이지 내용을 분석해
잠재 고객·투자자·채용 후보를 위한 짧은 브로셔를 만드는 어시스턴트입니다.
코드 블록 없이 마크다운으로만 답하세요.
"""
```

```python
def stream_brochure(company_name, url, model):
    yield ""
    prompt = f"{company_name} 회사 브로셔를 만들어 줘. 랜딩 페이지 내용은 다음과 같아:\n"
    prompt += fetch_website_contents(url)
    if model=="GPT":
        result = stream_gpt(prompt)
    elif model=="gemini":
        result = stream_gemini(prompt)
    else:
        raise ValueError("알 수 없는 모델")
    yield from result
```

```python
name_input = gr.Textbox(label="회사 이름:")
url_input = gr.Textbox(label="랜딩 페이지 URL (http:// 또는 https:// 포함)")
model_selector = gr.Dropdown(["GPT", "gemini"], label="모델 선택", value="GPT")
message_output = gr.Markdown(label="결과:")

view = gr.Interface(
    fn=stream_brochure,
    title="브로셔 생성기", 
    inputs=[name_input, url_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Hugging Face", "https://huggingface.co", "GPT"],
            ["Edward Donner", "https://edwarddonner.com", "gemini"]
        ], 
    flagging_mode="never"
    )
view.launch()
```
